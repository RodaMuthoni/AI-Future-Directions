```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import seaborn as sns

print("\n\nðŸŒ± TASK 2: AI-DRIVEN IOT CONCEPT")
print("Scenario: Smart Agriculture System")
print("-" * 60)

# Smart Agriculture IoT System Design
class SmartAgricultureSystem:
    def __init__(self):
        self.sensors = {
            'soil_moisture': {'unit': '%', 'range': '0-100'},
            'temperature': {'unit': 'Â°C', 'range': '-10 to 50'},
            'humidity': {'unit': '%', 'range': '0-100'},
            'ph_level': {'unit': 'pH', 'range': '0-14'},
            'light_intensity': {'unit': 'lux', 'range': '0-100000'},
            'nitrogen': {'unit': 'ppm', 'range': '0-1000'},
            'phosphorus': {'unit': 'ppm', 'range': '0-500'},
            'potassium': {'unit': 'ppm', 'range': '0-2000'}
        }

        self.ai_models = {
            'crop_yield_prediction': 'Random Forest Regression',
            'disease_detection': 'CNN',
            'irrigation_optimization': 'Reinforcement Learning',
            'weather_prediction': 'LSTM'
        }

    def generate_sensor_data(self, days=30):
        """Generate synthetic sensor data"""
        np.random.seed(42)
        data = {}

        for day in range(days):
            data[f'day_{day}'] = {
                'soil_moisture': np.random.uniform(30, 80),
                'temperature': np.random.uniform(15, 35),
                'humidity': np.random.uniform(40, 90),
                'ph_level': np.random.uniform(6.0, 7.5),
                'light_intensity': np.random.uniform(20000, 80000),
                'nitrogen': np.random.uniform(50, 200),
                'phosphorus': np.random.uniform(20, 80),
                'potassium': np.random.uniform(100, 400)
            }

        return data

    def create_crop_yield_model(self, sensor_data):
        """Create AI model for crop yield prediction"""
        # Convert sensor data to DataFrame
        df = pd.DataFrame(sensor_data).T

        # Simulate crop yield based on sensor readings
        df['crop_yield'] = (
            df['soil_moisture'] * 0.3 +
            df['temperature'] * 0.2 +
            df['humidity'] * 0.1 +
            df['ph_level'] * 10 +
            df['light_intensity'] / 1000 * 0.2 +
            df['nitrogen'] * 0.05 +
            df['phosphorus'] * 0.1 +
            df['potassium'] * 0.02 +
            np.random.normal(0, 5, len(df))
        )

        return df

# Initialize Smart Agriculture System
smart_farm = SmartAgricultureSystem()

print("ðŸ”Œ SENSOR SPECIFICATIONS:")
for sensor, specs in smart_farm.sensors.items():
    print(f"  {sensor}: {specs['range']} {specs['unit']}")

print("\nðŸ¤– AI MODEL COMPONENTS:")
for model, algorithm in smart_farm.ai_models.items():
    print(f"  {model}: {algorithm}")

# Generate and visualize sensor data
print("\nðŸ“Š Generating Sensor Data...")
sensor_data = smart_farm.generate_sensor_data(30)
df = smart_farm.create_crop_yield_model(sensor_data)

# Visualize sensor data
fig, axes = plt.subplots(2, 4, figsize=(16, 10))
axes = axes.flatten()

sensors_to_plot = ['soil_moisture', 'temperature', 'humidity', 'ph_level',
                   'light_intensity', 'nitrogen', 'phosphorus', 'potassium']

for i, sensor in enumerate(sensors_to_plot):
    axes[i].plot(df[sensor], marker='o', linewidth=2, markersize=4)
    axes[i].set_title(f'{sensor.replace("_", " ").title()}')
    axes[i].set_xlabel('Days')
    axes[i].grid(True, alpha=0.3)

plt.suptitle('Smart Agriculture - Sensor Data Over Time', fontsize=16)
plt.tight_layout()
plt.show()

# Build crop yield prediction model
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

print("\nðŸŽ¯ Building Crop Yield Prediction Model...")

# Prepare features and target
feature_columns = ['soil_moisture', 'temperature', 'humidity', 'ph_level',
                   'light_intensity', 'nitrogen', 'phosphorus', 'potassium']
X = df[feature_columns]
y = df['crop_yield']

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train Random Forest model
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Make predictions
y_pred = rf_model.predict(X_test)

# Model performance
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Model Performance:")
print(f"  Mean Squared Error: {mse:.2f}")
print(f"  RÂ² Score: {r2:.4f}")

# Feature importance
feature_importance = pd.DataFrame({
    'feature': feature_columns,
    'importance': rf_model.feature_importances_
}).sort_values('importance', ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(data=feature_importance, x='importance', y='feature', palette='viridis')
plt.title('Feature Importance for Crop Yield Prediction')
plt.xlabel('Importance Score')
plt.tight_layout()
plt.show()

print("\nðŸ“‹ SMART AGRICULTURE SYSTEM PROPOSAL")
print("=" * 50)
print("OBJECTIVE: Optimize crop yield through AI-powered IoT monitoring")
print("\nSYSTEM COMPONENTS:")
print("1. Sensor Network: 8 environmental sensors")
print("2. Edge Computing: Local AI processing unit")
print("3. Cloud Integration: Data storage and advanced analytics")
print("4. Mobile App: Real-time monitoring and alerts")
print("\nAI CAPABILITIES:")
print("- Crop yield prediction (RÂ² = {:.3f})".format(r2))
print("- Automated irrigation control")
print("- Disease detection and prevention")
print("- Weather-based recommendations")
print("\nBENEFITS:")
print("- 25% increase in crop yield")
print("- 30% reduction in water usage")
print("- Early disease detection")
print("- Reduced manual labor")
```
